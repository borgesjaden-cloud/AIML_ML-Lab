```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,roc_auc_score, confusion_matrix, classification_report
import pandas as pd
```

## Load Dataset

```python
data = load_breast_cancer()
X = data.data
y = data.target
print("Classes:", data.target_names)
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head()
```

```python
df.info()
```

```python
df.shape
```

## Split The Data

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,y,test_size=0.3,random_state=0
)
```

## Logistic Regression

```python
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)
print("Logistic Regression")
print(f"Accuracy: {accuracy_score(y_test, pred_lr)}")
pred_lr_train = lr.predict(X_train)
print(f"Training Accuracy: {accuracy_score(y_train, pred_lr_train)}")
```

## Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
pred_dt = dt.predict(X_test)
print("Decicion Tree")
print(f"Accuracy: {accuracy_score(y_test, pred_dt)}")
pred_dt_train = dt.predict(X_train)
print(f"Training Accuracy: {accuracy_score(y_train, pred_dt_train)}")
```

## KNN

```python
knn= KNeighborsClassifier()
knn.fit(X_train, y_train)
pred_knn = knn.predict(X_test)
print("KNN")
print(f"Accuracy: {accuracy_score(y_test, pred_knn)}")
pred_knn_train = knn.predict(X_train)
print(f"Training Accuracy: {accuracy_score(y_train, pred_knn_train)}")
```

## Naive Bayes

```python
NB = GaussianNB()
NB.fit(X_train, y_train)
pred_NB = NB.predict(X_test)
print("Naive Bayes")
print(f"Accuracy: {accuracy_score(y_test, pred_NB)}")
pred_NB_train = NB.predict(X_train)
print(f"Training Accuracy: {accuracy_score(y_train, pred_NB_train)}")
```

## MLP

```python
MLP = MLPClassifier()
MLP.fit(X_train, y_train)
pred_MLP = MLP.predict(X_test)
print("MLP")
print(f"Accuracy: {accuracy_score(y_test, pred_MLP)}")
pred_MLP_train = MLP.predict(X_train)
print(f"Training Accuracy: {accuracy_score(y_train, pred_MLP_train)}")
```
